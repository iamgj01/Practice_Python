```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Legend-10/ML_CSV/main/covid.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
df1 = ['Last_Update','Lat','Long_','UID','ISO3']
```

```python
df.drop(df1,axis=1,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df2 = ['Recovered','Active','People_Hospitalized','Hospitalization_Rate','People_Tested','Mortality_Rate', 'Country_Region']
```

```python
df.drop(df2,axis=1,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.drop('FIPS', axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Incident_Rate'].fillna(df['Incident_Rate'].mean(),inplace=True)
df['Total_Test_Results'].fillna(df['Total_Test_Results'].mean(),inplace=True)
df['Case_Fatality_Ratio'].fillna(df['Case_Fatality_Ratio'].mean(),inplace=True)
df['Testing_Rate'].fillna(df['Testing_Rate'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

```python
df['Date'].fillna(df['Date'].ffill(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```
